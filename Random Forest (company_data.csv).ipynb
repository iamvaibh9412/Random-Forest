# Assignment - Random Forest (Company_data.csv)

```python
# importing required libraries

import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
company = pd.read_csv('Company_Data.csv')
company
```

```python
#considering sales >= 9 as 'Good' & < 9 as 'Risky'

sales_val=[]
for value in company['Sales']:
    if value >= 9:
        sales_val.append('Good')
    else :
        sales_val.append('Risky')
company['sales_val'] = sales_val
```

```python
company
```

```python
company_data = company.drop(['Sales'], axis=1)
company_data
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
company_data['ShelveLoc']= label_encoder.fit_transform(company_data['ShelveLoc'])
company_data['Urban']= label_encoder.fit_transform(company_data['Urban'])
company_data['US']= label_encoder.fit_transform(company_data['US'])
```

```python
company_data
```

```python
array = company_data.values
x= array[:,0:10]
y= array[:,10:]
```

```python
x
```

```python
y
```

```python
kfold = KFold( n_splits=6, random_state=None)
model= RandomForestClassifier(n_estimators=100,max_features=6)
results= cross_val_score(model, x, y.ravel(), cv=kfold)     # .ravel will convert array shape to (n, ) i.e 1d array
print("The Accuracy is :",results.mean())
```

### Accuracy = 83%
