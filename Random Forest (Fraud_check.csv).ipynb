# Assignment - Random Forest (Fraud_check.csv)

```python
# importing required libraries

import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
fraud_df= pd.read_csv('Fraud_check.csv')
fraud_df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
fraud_df['Undergrad']=label_encoder.fit_transform(fraud_df['Undergrad'])
fraud_df['Marital.Status']= label_encoder.fit_transform(fraud_df['Marital.Status'])
fraud_df['Urban'] = label_encoder.fit_transform(fraud_df['Urban'])
```

```python
fraud_df
```

```python
tax_val=[]
for value in fraud_df['Taxable.Income']:
    if value > 30000:
        tax_val.append('Good')
    else :
        tax_val.append('Risky')
fraud_df['tax_val']= tax_val
```

```python
fraud_df
```

```python
fraud_data= fraud_df.drop(['Taxable.Income'], axis=1)
fraud_data
```

```python
array =  fraud_data.values
x = array[:, 0:5]
y= array[:,5:]
```

```python
x
```

```python
y
```

```python
num_trees = 100
max_features = 4
kfold = KFold(n_splits=10, random_state=None)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, x,y.ravel(), cv=kfold)         # .ravel will convert array shape to (n, ) i.e 1d array
print("The Accuracy is :", results.mean())
```

#### The Accuracy is 74%
